**Punto 1:** Cargar y visualizar la imagen

```python
import matplotlib.pyplot as plt
from scipy import ndimage
import cv2
import numpy as np

global imagen 
imagen = "cameraman.ppm"

# Cargar la imagen en blanco y negro
orgImg = cv2.imread(imagen, cv2.IMREAD_GRAYSCALE) #recibe la ruta de la imagen

# Visualizar la imagen original
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(orgImg, cmap='gray')

plt.show()
```

**Punto 2:**
Definir los filtros

```python
# Crear matrices de los filtros a utilizar

h1 = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0]
])

h2 = (1/9) * np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

h3 = (1/273) * np.array([
    [1, 4, 7, 4, 1],
    [4, 16, 26, 16, 4],
    [7, 26, 41, 26, 7],
    [4, 16, 26, 16, 4],
    [1, 4, 7, 4, 1]
])

h4 = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

h5 = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

h6 = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
])

h7 = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])

h8 = np.array([
    [-2, -1, 0],
    [-1, 1, 1],
    [0, 1, 2]
])
```

Aplique y visualice los filtros


```python
# Aplicar el filtro H1
filteredImg = cv2.filter2D(orgImg, cv2.CV_32F, h1)  # Usar orgImg en lugar de imagen

# Visualizar la imagen original y la imagen filtrada en subplots
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.title('Imagen Original')
plt.imshow(orgImg, cmap='gray')  # Mostrar la imagen cargada correctamente

plt.subplot(1, 2, 2)
plt.title('Imagen Filtrada con h1')
plt.imshow(filteredImg, cmap='gray')

plt.show()
```
