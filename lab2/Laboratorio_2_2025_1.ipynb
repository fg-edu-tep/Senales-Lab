## **1. Definición de funciones básicas: rampa, escalón unitario e impulso unitario.**

```python
import numpy as np
import matplotlib.pyplot as plt

def ramp(t, delay):
    """
    ramp: función rampa
    u = (t + delay) si t >= -delay
    u = 0 en otro caso
    """
    u = np.where(t >= -delay, t + delay, 0)
    return u

def unitStep(t, delay):
    """
    unitStep: función escalón unitario
    u = 1 si t >= -delay
    u = 0 en otro caso
    """
    u = np.where(t >= -delay, 1, 0)
    return u

def unit_impulse(delay):
    # Crear una secuencia de tiempo discreta
    n = np.arange(-10, 11)

    # Crear la señal del impulso unitario con el corrimiento
    impulse = np.zeros_like(n)
    impulse[n == -delay] = 1

    # Graficar la señal del impulso unitario
    plt.stem(n, impulse)
    plt.xlabel('n')
    plt.ylabel('Amplitud')
    plt.title(f'Impulso Unitario con Corrimiento ({delay})')
    plt.xticks(np.arange(min(n), max(n)+1, 1))  
    plt.grid(True)
    plt.show()

    return 
```

## **2. Ejemplos.**

### - Escalón unitario: $u(-t-3)$

```python
# Se genera un vector t con los valores de tiempo
t = np.linspace(-10, 10, 1000)

# Se define el valor de delay para el escalón
delay_escalon = -3

#  Se calcula la función escalón
u_escalon = unitStep(-t, delay_escalon)

# Se grafica la función escalón
plt.plot(t, u_escalon, label='Escalón') # Se grafica la función escalón
plt.xlabel('Tiempo') # Se etiqueta el eje x
plt.ylabel('Valor') # Se etiqueta el eje y
plt.title('Función Escalón') # Se coloca un título a la gráfica
plt.xticks(np.arange(min(t), max(t)+1, 1))  # Se ajustan los valores del eje x
plt.legend() # Se coloca la leyenda
plt.grid(True) # Se activa la malla
plt.show() # Se muestra la gráfica
```

### - Rampa: $r(t-3)$

```python
# Se genera un vector t con los valores de tiempo
t = np.linspace(-10, 10, 1000)

# Se define el valor de delay para el escalón
delay_rampa = -3

#  Se calcula la función rampa
u_rampa = ramp(t, delay_rampa)

# Se grafica la función rampa
plt.plot(t, u_rampa, label='Rampa') # Se grafica la función rampa
plt.xlabel('Tiempo') # Se etiqueta el eje x
plt.ylabel('Valor') # Se etiqueta el eje y
plt.title('Función Rampa') # Se coloca el título de la gráfica
plt.xticks(np.arange(min(t), max(t)+1, 1))  # Se ajustan los valores del eje x
plt.legend() # Se coloca la leyenda
plt.grid(True)  # Se activa la cuadrícula
plt.show()  # Se muestra la gráfica
```

### - Impulso unitario $\delta(t-3)$

```python
# Se define el valor de delay para el impulso
delay_impulso = -3

#  Se calcula y grafica la función impulso
unit_impulse(delay_impulso)
```
